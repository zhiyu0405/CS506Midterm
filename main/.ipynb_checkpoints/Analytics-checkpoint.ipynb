```python
import pandas as pd
df = pd.read_csv('../data/X_train.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df[['ProductId', 'UserId']].groupby('ProductId').count().nlargest(50, 'UserId').plot(kind = 'bar')
plt.show()
```

```python
df[['UserId', 'ProductId']].groupby('UserId').count().nlargest(50, 'ProductId').plot(kind = 'bar')
plt.show()
```

```python
df[['Score', 'UserId']].groupby('Score').count().plot(kind = 'bar')
plt.show()
```

```python
gp = df[['UserId', 'Helpfulness']].groupby('UserId').filter(lambda x: len(x) > 50)
```

```python
gp.groupby('UserId').mean().nlargest(50, 'Helpfulness').plot(kind = 'bar')
plt.show()
```
